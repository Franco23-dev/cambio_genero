```python
#Importacion de librerias utilizadas
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Abrimos el dataset
datos = pd.read_csv("cambio_genero_provincia_genero_edad_mayo_2024-copia.csv")
```

```python
#Eliminamos la columna 'provincia_id' por no tener relevancia en el analisis
datos_clean = datos.drop(columns="provincia_id")
```

```python
#Revisamos que informacion que contiene
datos_clean.head()
```

```python
#Guardamos los dataframes 
df = pd.DataFrame(datos_clean)
```

```python
#Separamos por grupos utilizando la funcion 'groupby'
df_genero = df.groupby("genero")
df_provincia = df.groupby("provincia_nombre")
df_edades = df.groupby("edad_actual_grandes_grupos")
```

```python
#Mostramos los grupos del genero 'X'
df_genero.get_group("X")
```

```python
#Mostramos los grupos de la provincia de 'Chubut'
df_provincia.get_group("Chubut")
```

```python
#Mostramos por grupo de '18 a 29' años
df_edades.get_group("18 a 29")
```

```python
#Sacamos el promedio de cada edad actual en grandes grupos
df_edades["cantidad"].mean()
```

```python
#Sacamos el promedio por genero
df_genero["cantidad"].mean()
```

```python
#Sacamos el promedio por 'provincia'
df_provincia["cantidad"].mean()
```

```python
#Contamos cuanto hay por cada genero
df_genero.count()
```

```python
#Contamos cuantos hay por edad actual y grandes grupos
df_edades.count()
```

```python
#Contamos cuantos hay por provincias
df_provincia.count()
```

```python
#Sacamos cuanto es el promedio de edad actual y grandes grupos por cada provincia 
df.pivot_table(index="provincia_nombre", columns="edad_actual_grandes_grupos",values="cantidad",aggfunc="mean")
```

```python
#Sacamos el maximo de cada uno de las provincias por genero
df.pivot_table(index="provincia_nombre", columns="genero", values="cantidad", aggfunc="max")
```

```python
#Comparamos cuanta frecuencia hay entre cada genero por edad actual y grandes grupos
res = pd.crosstab(df["edad_actual_grandes_grupos"], df["genero"])
print(res)
```
